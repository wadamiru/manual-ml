# Logistic Regression (Binary Classification)

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate data

the model will classify points in 2D into 2 seperate classes. \
#### class 0: cluster centered around (-2,-2)
- $x ∼ \mathcal{N}(\mu=(-2,-2), \Sigma=I)$

#### class 1: cluster centered around (2,2)
- $x∼\mathcal{N}(\mu=(2,2), \Sigma=I)$

Unlike classical linear classification, which outputs a hard label using a step function 
$$
y = 
\begin{cases} 
1 & \text{if } w \cdot x + b > 0 \\ 
0 & \text{otherwise} 
\end{cases},
$$
logistic regression outputs probabilities via the sigmoid function 
$$
\sigma(z) = \frac{1}{1 + e^{-z}}, \quad z = w \cdot x + b.
$$
the decision boundary is still linear.

#### *logistic regression: probabilistic output*
instead of a hard label, logistic regression outputs the probability that a point belongs to class 1:
$$P(y=1|x,\theta)=\sigma(z)$$
similarly, the probability of belonging to class 0:
$$P(y=0|x, \theta)=1-\sigma(z)$$

```python
# generate data shape
# x: (500,2), 250 from each class
# y: (500,), correct class (0/1) for all 500 x

n = 250 # n samples
```

```python
def make_data(n=100):
    c0 = np.random.randn(n,2) + np.array([-2,-2])
    c1 = np.random.randn(n,2) + np.array([2,2])

    x = np.vstack([c0,c1])
    y = np.hstack([np.zeros(n), np.ones(n)])

    return x,y
```

```python
x,y = make_data(n)
print(x.shape)
print(y.shape)
```

```python
def plot_data(x,y,label="training"):
    c0 = x[y==0]
    c1 = x[y==1]

    plt.scatter(c0[:,0], c0[:,1], label="class 0")
    plt.scatter(c1[:,0], c1[:,1], label="class 1")
    plt.legend()
    plt.title(f"dataset: {label}")
    plt.show()
```

```python
# visualise data
plot_data(x,y)
```

## Model implementation

```python
# sigmoid activation function
def sig(z):
    # clip pre-activation to avoid overflow
    z = np.clip(z, -500, 500)
    return 1/(1+np.exp(-z))
```

```python
# forward pass
def forward(x, w, b):
    # x: (nsamples, nclasses) | w,b: (nclasses,)
    z = x @ w + b
    p = sig(z)
    return p
```

## Loss function

Given $N$ independent training examples $(x_i, y_i)$, the likelihood of the parameters $w, b$ is:

$$
L(w,b) = \prod_{i=1}^{N} P(y_i \mid x_i) 
       = \prod_{i=1}^{N} \sigma(z_i)^{y_i} (1 - \sigma(z_i))^{1 - y_i},
$$

where $z_i = w \cdot x_i + b$, and $\sigma(z) = \frac{1}{1 + e^{-z}}$ is the sigmoid function.
If $y^i = 0$, the probability of class $0$ ($(1 - \sigma(z_i))$) will be counted, whereas if $y^i = 1$, the probability of class $1$ ($\sigma(z_i)$) will be counted.

#### Log-likelihood:

$$
\ell(w,b) = \log L(w,b) 
          = \sum_{i=1}^{N} \Big[ y_i \log \sigma(z_i) + (1 - y_i) \log (1 - \sigma(z_i)) \Big]
$$

#### Negative log-likelihood / cross-entropy loss:

$$
\mathcal{L}(w,b) = - \ell(w,b) 
                  = - \sum_{i=1}^{N} \Big[ y_i \log \sigma(z_i) + (1 - y_i) \log (1 - \sigma(z_i)) \Big]
$$

#### Gradient with respect to $w$:

$$
\frac{\partial \mathcal{L}}{\partial w} = \sum_{i=1}^{N} (\sigma(z_i) - y_i) x_i
$$

```python
# we minimise binary-cross entropy
# L = -1/n . sum { y.log p + (1-y)log(1-p) }
def loss(y,p):
    eps = 1e-15 # noise
    # clip probs before taking log
    p = np.clip(p, eps, 1-eps)
    # terms
    t1 = y*np.log(p)
    t2 = (1-y)*np.log(1-p)

    return -np.mean(t1 + t2)
```

## Gradient derivation

```python
# dL/d{w,b} = dL/dp . dp/dz . dz/d{w,b} = dL/dz . dz/d{w,b}
# dz = p-y | dw = (p-y)x | db = (p-y)
# matrix versions: dw = mean{(p-y)x}, db = mean{(p-y)} | mean = (1/n)sum(_)
def grads(x,y,p):
    # x: (nsamples, nclasses), y: (nsamples,), p: (nsamples,)
    dz = p - y
    dw = (1/x.shape[0]) * x.T @ dz
    db = np.mean(dz)

    return dw,db
```

## Training the model

```python
def train(x, y, lr=0.1, steps=1000):
    # parameter setup
    w = np.zeros(x.shape[1]) # nclasses
    b = 0
    y = y.reshape(-1)
    
    Ls = [] # losses
    for s in range(steps):
        p = forward(x,w,b)
        Ls.append(loss(y,p))

        dw, db = grads(x,y,p)
        # gradient descent
        w -= lr * dw
        b -= lr * db

    return w,b,Ls
```

```python
# train the model
w,b,Ls = train(x,y)
```

## Analysis

```python
def predict(x,w,b):
    p = forward(x,w,b)
    return (p >= 0.5).astype(int)
```

```python
# create a new dataset for testing
xte, yte = make_data(100)
```

```python
# get predictions
ps = predict(xte,w,b)

# calculate Accuracy
acc = (ps==yte).mean()
print("Accuracy (testing data):",acc)
```

```python
# visualise descision boundary of data
def plot_bound(x, y, w, b, label="training data"): # label: training or testing

    class0 = x[y==0]
    class1 = x[y==1]

    plt.scatter(class0[:,0], class0[:,1])
    plt.scatter(class1[:,0], class1[:,1])

    x_vals = np.linspace(-5,5,100)
    y_vals = -(w[0]*x_vals + b)/w[1]

    plt.plot(x_vals, y_vals, color="black")

    plt.title(f"Decision Boundary: {label}")
    plt.show()
```

```python
# visualise testing data & descision boundary
plot_data(xte,yte,"testing")
plot_bound(xte,yte,w,b, "testing data")
```

```python
# descision bound for training data
plot_bound(x,y,w,b)
```
